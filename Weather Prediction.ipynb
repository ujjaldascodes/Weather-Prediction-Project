```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('datafile_Worksheet.csv')
```

```python
data.drop(['JAN-FEB','MAR-MAY','JUN-SEP','OCT-DEC'],axis=1,inplace=True)
```

```python
data.info()
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
plot=sns.pairplot(data)
plt.show()
```

```python
sns.distplot(data['ANNUAL'])#Plot according to the column
plt.show()
```

```python
sns.heatmap(data.corr(),linecolor='white',linewidths=1)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True,linecolor='white',linewidths=0.00001)#shows the actual values
plt.show()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Train the data part
x_train, x_test, y_train, y_test=train_test_split(x, y,test_size=0.20,random_state=200)
```

```python
#Createing the model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

```

```python
model.fit(x_train,y_train)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
prediction=model.predict(x_test)
print(prediction)
```

```python
predicteddata=pd.DataFrame(y_test,prediction,columns=['Collected Data vs Predicted Data'])
predicteddata
```

```python
plt.scatter(y_test,prediction,color='red',edgecolor='black',linewidths=1.5)
plt.show()
```

```python
sns.distplot((y_test-prediction),bins=10,rug=True,color='red')
plt.show()
```

```python
# Map to upper,lower, and diagonal
z=sns.PairGrid(predicteddata,height=6.0,aspect=1.7)
z.map_diag(plt.hist)
z.map_upper(plt.scatter)
z.map_lower(sns.kdeplot)
plt.show()
```

```python
from sklearn import metrics
#for counting the mathematical problems
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,prediction))
print('MSE:', metrics.mean_squared_error(y_test,prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

```python
metrics.explained_variance_score(y_test,prediction)#The accuracy of the project and 1.0 is the highest possible value that is 100%
```
